```python
import numpy as np
import matplotlib.pyplot as plt
```

# Regression to the mean

```python
n = 1000000
k = 5
X = np.random.rand(n, k).max(axis=1)
```

```python
Xi = np.random.rand(n)
```

$X_i \sim p_i(x)$ where $p_i(x)$ is the p.d.f. of $X_i$. The c.d.f. of $X_i$ is $F_i(x) = P(X_i < x)$.

$X = \max{X_i}$. 

Let $F(x)$ be the c.d.f. of $X$, that is, $F(x) = P(X < x)$.

Then:

$$
\begin{align*}
F(x) &= P(X < x) \\
&= P((X_1 < x) \wedge (X_2 < x) \wedge \cdots \wedge (X_n < x)) \\
&= P(X_1 < x) P(X_2 < x) \cdots P(X_n < x) \\
&= \prod_{i=1}^{n} F_i(x)
\end{align*}
$$

Therefore:

$$
F(x) = \prod_{i=1}^{n} F_i(x)
$$

```python
# Comparing the pdf of X and Xi.
plt.hist(X, density=True, bins=500, alpha=0.5, label=r'$\text{pdf}(X)$')
plt.hist(Xi, density=True, bins=500, alpha=0.5, label=r'$\text{pdf}(X_i)$')
plt.xlabel(r'$x$')
plt.legend()
plt.show()
```
