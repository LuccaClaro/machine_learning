```python
import numpy as np
```

```python
X = np.array([[1, 2], [3, 4], [5, 6]])
X
```

```python
U, S, Vt = np.linalg.svd(X)
```

```python
U, S, Vt
```

```python
(U.T @ U).round(2)
```

```python
(Vt @ Vt.T).round(2)
```

```python
S_full = np.vstack([np.diag(S), np.zeros((1, 2))])
S_full
```

```python
U, S_full, Vt
```

```python
U @ S_full @ Vt
```

```python
U[:,:2] @ np.diag(S) @ Vt
```

```python
# Leia uma imagem png e converta para escala de cinza
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('image.png')
X = img.mean(axis=2)
plt.imshow(X, cmap='gray')
```

```python
X.shape
```

```python
U, S, Vt = np.linalg.svd(X)
```

```python
plt.plot(S)
```

```python
U_aprox = U[:, :40]
S_aprox = S[:40]
Vt_aprox = Vt[:40, :]
```

```python
U_aprox.shape, S_aprox.shape, Vt_aprox.shape
```

```python
X_aprox = U_aprox @ np.diag(S_aprox) @ Vt_aprox
```

```python
plt.imshow(X_aprox, cmap='gray')
```
